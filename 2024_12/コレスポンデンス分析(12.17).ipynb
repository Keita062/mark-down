```python
import mca
```

```python
mca_counts = mca.MCA(Int_64, benzecri=False)

rows = mca_counts.fs_r(N=2)
cols = mca_counts.fs_c(N=2)
```

```python
plt.figure(figsize=(15, 8))

plt.scatter(rows[:, 0], rows[:, 1], marker="None")
labels = df.index
for label, x, y in zip(labels, rows[:, 0], rows[:, 1]):
  plt.annotate(label, xy=(x, y), c="b")
```

```python
# プロットの準備
plt.figure(figsize=(15, 8))

# 行スコアをプロット (ラベルはインデックス名)
row_labels = Int_64.index
plt.scatter(rows[:, 0], rows[:, 1], marker="o", label="Rows", c="blue")
for label, x, y in zip(row_labels, rows[:, 0], rows[:, 1]):
    plt.annotate(label, xy=(x, y), color="blue", fontsize=10)

# 列スコアをプロット (ラベルはカラム名)
col_labels = Int_64.columns
plt.scatter(cols[:, 0], cols[:, 1], marker="s", label="Columns", c="red")
for label, x, y in zip(col_labels, cols[:, 0], cols[:, 1]):
    plt.annotate(label, xy=(x, y), color="red", fontsize=10)

# プロットの調整
plt.axhline(0, color="gray", linestyle="--", linewidth=0.5)
plt.axvline(0, color="gray", linestyle="--", linewidth=0.5)
plt.title("MCA Biplot", fontsize=16)
plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

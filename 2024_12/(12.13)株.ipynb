```python
# 条件に基づいてデータを抽出
Toyota_time = Toyota_time[(Toyota_time['Date'] >= '2004-01-04') & (Toyota_time['Date'] <= '2024-12-12')]
Aisin_time = Aisin_time[(Aisin_time['Date'] >= '2000-01-04') & (Aisin_time['Date'] <= '2024-12-12')]
Fujitu_time = Fujitu_time[(Fujitu_time['Date'] >= '2000-01-04') & (Fujitu_time['Date'] <= '2024-12-12')]
Hitati_time = Hitati_time[(Hitati_time['Date'] >= '2000-01-04') & (Hitati_time['Date'] <= '2024-12-12')]
Hondagiken_time = Hondagiken_time[(Hondagiken_time['Date'] >= '2000-01-04') & (Hondagiken_time['Date'] <= '2024-12-12')]
Nissan_time = Nissan_time[(Nissan_time['Date'] >= '2000-01-04') & (Nissan_time['Date'] <= '2024-12-12')]
Softbank_time = Softbank_time[(Softbank_time['Date'] >= '2000-01-04') & (Softbank_time['Date'] <= '2024-12-12')]
Sony_time = Sony_time[(Sony_time['Date'] >= '2000-01-04') & (Sony_time['Date'] <= '2024-12-12')]
Suzuki_time = Suzuki_time[(Suzuki_time['Date'] >= '2000-01-04') & (Suzuki_time['Date'] <= '2024-12-12')]
```

```python
X = Toyota_time["Date"]
X2 = Fujitu_time["Date"]
Y = Toyota_time["Close"]
Y2 = Fujitu_time["Close"]
width = 15
height = 9  

Blue = "blue"
Red = "red"

plt.figure(figsize=(width, height))

# Toyota_Closeのプロット
plt.plot(X, Y, color=Red, label="Toyota_Close")

# Fujitu_Closeのプロット
plt.plot(X2, Y2, color=Blue, label="Fujitu_Close")

# ラベルやタイトル、y軸の目盛りを設定
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.title("Toyota vs Fujitu Close Prices Over Time")
plt.yticks(range(0, 5001, 500))
plt.grid(True)

# 凡例を追加
plt.legend()

# グラフを表示
plt.show()
```

次元数が多いため、主成分分析(PCA)を行い、次元数を削除する.

```python
# 主成分分析のためのライブラリをセット
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
# 標準化
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X.head()
```

```python
#　標準化されたデータ
X_scaled
```

```python
pca = PCA()
X_pca = pca.fit_transform(X_scaled)
```

```python
# 次元削除されたデータの形を直接確認
print("変換後のデータ:\n", X_pca)
```

```python
# 線形結合の係数の確認(主成分の係数)
print("主成分の係数:\n", pca.components_)
```

```python
# 累積分散比率の確認
cumulative_variance_ratio = np.cumsum(pca.explained_variance_ratio_)
print("累積分散比率:", cumulative_variance_ratio)
```

```python
# 必要な次元数を選択（例：95%以上の分散を保持）
n_components = np.argmax(cumulative_variance_ratio >= 0.95) + 1
print(f"95%以上の分散を保持するための主成分数: {n_components}")
```

```python
# 再度PCAを実施して次元削減
pca = PCA(n_components=n_components)
X_reduced = pca.fit_transform(X_scaled)
```

```python
# 主成分の係数を確認
loadings = pd.DataFrame(pca.components_, columns=X.columns,
                        index=[f'PC{i+1}' for i in range(pca.n_components_)])
loadings
```

```python
import scipy.stats as stats
```

```python
Hondagiken_time["Month"] = Hondagiken_time["Date"].dt.to_period("M")
```

```python
Hondagiken_time.head()
```

```python
Monthly_avg_stock_Honda = Hondagiken_time.groupby('Month')['Adj Close'].mean().reset_index()
Monthly_avg_stock_Honda['Month'] = Monthly_avg_stock_Honda['Month'].dt.to_timestamp()
```

```python
merged_data = pd.merge(
    Search_Data,  # Search_Data を左側（left）に
    Monthly_avg_stock_Honda,  # monthly_avg_stock を右側（right）に
    left_on='Date',  # Search_Data の 'Date' を基準に
    right_on='Month',  # monthly_avg_stock の 'Month' を基準に
    how='inner'  # inner join を使用
)
```

```python
merged_data.head()
```

```python
merged_data = merged_data.drop(columns=['Month', 'Toyota', 'Fujitu', 'Hitati', 'Nissan', 'Softbank', 'Sony', 'Suzuki'])
merged_data = merged_data.rename(columns={'Adj Close': 'Hondagiken_Adj Close'})
merged_data.head()
```

```python
# ピアソンの相関係数を計算
corr_coefficient_personr, p_value_personer = stats.pearsonr(merged_data["Honda"], merged_data["Hondagiken_Adj Close"])

# スピアマンの相関係数を計算
corr_coefficient_spearmanr,p_value_spearmanr = stats.spearmanr(merged_data["Honda"], merged_data["Hondagiken_Adj Close"])

# 結果を表示
print(f"ピアソン相関係数: {corr_coefficient_personr}")
print(f"ピアソンp値: { p_value_personer}")

print(f"スピアマン相関係数: {corr_coefficient_spearmanr}")
print(f"スピアマンp値: { p_value_spearmanr }")

# 帰無仮説と対立仮説の検定
alpha = 0.05  # 有意水準
if p_value_personer < alpha:
    print("ピアソンの場合、帰無仮説を棄却します。株価と検索数に相関があります。")
else:
    print("ピアソンの場合、帰無仮説を棄却できません。株価と検索数に相関はないと考えられます。")

if p_value_spearmanr < alpha:
   print("スピアマンの場合、帰無仮説を棄却します。株価と検索数に相関があります。")
else:
   print("スピアマンの場合、帰無仮説を棄却できません。株価と検索数に相関はないと考えられます。") 
```

```python
from sklearn.linear_model import LinearRegression
```

```python
import seaborn as sns
```

```python
Ex_var = merged_data["Honda"].values.reshape(-1, 1)  # 説明変数（X）
Ob_var = merged_data["Hondagiken_Adj Close"].values  # 目的変数（Y）
```

```python
# 単回帰分析のモデルを作成
model = LinearRegression()
model.fit(Ex_var, Ob_var)
```

```python
# 回帰直線の予測値を計算
Y_pred = model.predict(Ex_var)
```

```python
# 回帰分析の結果
slope = model.coef_[0]
intercept = model.intercept_
r_squared = model.score(Ex_var, Ob_var)
```

```python
print(f"回帰直線の傾き: {slope}")
print(f"回帰直線の切片: {intercept}")
print(f"決定係数 (R^2): {r_squared}")
```

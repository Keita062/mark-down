```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# 株式のティッカーシンボル（トヨタ自動車の場合）
ticker = "7203.T" # トヨタ
ticker2 = "7267.T" # 本田技研工業
ticker3 = "6501.T" # 日立製作所
ticker4 = "6758.T" # ソニーグループ
ticker5 = "7201.T" # 日産自動車
ticker6 = "9984.T" # ソフトバンク
ticker7 = "7359.T" # アイシン
ticker8 = "6702.T" # 富士通
ticker9 = "7269.T" # スズキ




# 特定の期間を指定してデータを取得
start_date = "2000-01-01"  # 開始日
end_date = "2024-01-01"    # 終了日
stock_data = yf.download(ticker, start=start_date, end=end_date)
stock_data2 = yf.download(ticker2, start=start_date, end=end_date)
stock_data3 = yf.download(ticker3, start=start_date, end=end_date)
stock_data4 = yf.download(ticker4, start=start_date, end=end_date)
stock_data5 = yf.download(ticker5, start=start_date, end=end_date)
stock_data6 = yf.download(ticker6, start=start_date, end=end_date)
stock_data7 = yf.download(ticker7, start=start_date, end=end_date)
stock_data8 = yf.download(ticker8, start=start_date, end=end_date)
stock_data9 = yf.download(ticker9, start=start_date, end=end_date)


# 株価の推移（終値）をプロット
plt.figure(figsize=(12, 6))
plt.plot(stock_data["Close"], label=f"{ticker} Close Price", color="blue")
plt.plot(stock_data2["Close"], label=f"{ticker2} Close Price", color="red")
plt.plot(stock_data3["Close"], label=f"{ticker3} Close Price", color="Black")
plt.plot(stock_data4["Close"], label=f"{ticker4} Close Price", color="green")
plt.plot(stock_data5["Close"], label=f"{ticker5} Close Price", color="purple")
plt.plot(stock_data6["Close"], label=f"{ticker6} Close Price", color="yellow")
plt.plot(stock_data7["Close"], label=f"{ticker7} Close Price", color="orange")
plt.plot(stock_data8["Close"], label=f"{ticker8} Close Price", color="#00AE95")
plt.plot(stock_data9["Close"], label=f"{ticker9} Close Price", color="#6F51A1")

plt.xlabel("Date")
plt.ylabel("Price (JPY)")
plt.legend()
plt.grid()
plt.show()
```

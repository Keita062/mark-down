```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\sk062\Documents\Data\Google Search\Search.csv")
Data = pd.DataFrame(df)
Data["Data"] = pd.to_datetime(Data["Data"])
```

```python
# 'month' 列を作成
Data['month'] = Data['Data'].dt.month
# 'date' 列を作成
Data['day'] = Data['Data'].dt.day

Data["week_number"] = Data["Data"].dt.isocalendar().week
```

```python
Data.head(20)
```

```python
Data_2019 = Data[(Data['Data'] >= '2019-12-15') & (Data['Data'] <= '2019-12-29')]
Data_2020 = Data[(Data['Data'] >= '2020-01-05') & (Data['Data'] <= '2020-12-27')]
Data_2021 = Data[(Data['Data'] >= '2021-01-03') & (Data['Data'] <= '2021-12-26')]
Data_2022 = Data[(Data['Data'] >= '2022-01-02') & (Data['Data'] <= '2022-12-25')]
Data_2023 = Data[(Data['Data'] >= '2023-01-01') & (Data['Data'] <= '2023-12-31')]
Data_2024 = Data[(Data['Data'] >= '2024-01-07') & (Data['Data'] <= '2024-12-15')]
```

```python
Data_2019
```

```python
# 2019年と2020年の検索数データ（例）
search_2019_AIR_JORDAN = Data_2019[["Data","AIR JORDAN","month","day","month","day","week_number"]] 
search_2020_AIR_JORDAN = Data_2020[["Data","AIR JORDAN","month","day","month","day","week_number"]]
search_2021_AIR_JORDAN = Data_2021[["Data","AIR JORDAN","month","day","month","day","week_number"]]
search_2022_AIR_JORDAN = Data_2022[["Data","AIR JORDAN","month","day","month","day","week_number"]]
search_2023_AIR_JORDAN = Data_2023[["Data","AIR JORDAN","month","day","month","day","week_number"]]
search_2024_AIR_JORDAN = Data_2024[["Data","AIR JORDAN","month","day","month","day","week_number"]]
```

```python
group_week_2019_AIR_JORDAN = search_2019_AIR_JORDAN.groupby("week_number")["AIR JORDAN"].sum()
group_week_2020_AIR_JORDAN = search_2020_AIR_JORDAN.groupby("week_number")["AIR JORDAN"].sum()
group_week_2021_AIR_JORDAN = search_2021_AIR_JORDAN.groupby("week_number")["AIR JORDAN"].sum()
group_week_2022_AIR_JORDAN = search_2022_AIR_JORDAN.groupby("week_number")["AIR JORDAN"].sum()
group_week_2023_AIR_JORDAN = search_2023_AIR_JORDAN.groupby("week_number")["AIR JORDAN"].sum()
group_week_2024_AIR_JORDAN = search_2024_AIR_JORDAN.groupby("week_number")["AIR JORDAN"].sum()
```

```python
group_week_2020_AIR_JORDAN = pd.DataFrame( group_week_2020_AIR_JORDAN )
group_week_2021_AIR_JORDAN = pd.DataFrame( group_week_2021_AIR_JORDAN )
group_week_2022_AIR_JORDAN = pd.DataFrame( group_week_2022_AIR_JORDAN )
group_week_2023_AIR_JORDAN = pd.DataFrame( group_week_2023_AIR_JORDAN )
group_week_2024_AIR_JORDAN = pd.DataFrame( group_week_2024_AIR_JORDAN )
group_week_2020_AIR_JORDAN.reset_index(inplace=True)
group_week_2021_AIR_JORDAN.reset_index(inplace=True)
group_week_2022_AIR_JORDAN.reset_index(inplace=True)
group_week_2023_AIR_JORDAN.reset_index(inplace=True)
group_week_2024_AIR_JORDAN.reset_index(inplace=True)
```

```python
week_2020_AIR_JORDAN = group_week_2020_AIR_JORDAN["week_number"]
week_2021_AIR_JORDAN = group_week_2021_AIR_JORDAN["week_number"]
week_2022_AIR_JORDAN = group_week_2022_AIR_JORDAN["week_number"]
week_2023_AIR_JORDAN = group_week_2023_AIR_JORDAN["week_number"]
week_2024_AIR_JORDAN = group_week_2024_AIR_JORDAN["week_number"]
count_2020_AIR_JORDAN = group_week_2020_AIR_JORDAN["AIR JORDAN"]
count_2021_AIR_JORDAN = group_week_2021_AIR_JORDAN["AIR JORDAN"]
count_2022_AIR_JORDAN = group_week_2022_AIR_JORDAN["AIR JORDAN"]
count_2023_AIR_JORDAN = group_week_2023_AIR_JORDAN["AIR JORDAN"]
count_2024_AIR_JORDAN = group_week_2024_AIR_JORDAN["AIR JORDAN"]
```

```python
plt.bar(week_2020_AIR_JORDAN, count_2020_AIR_JORDAN, width=1.0, edgecolor="black")


plt.xlabel( "week number" )
plt.ylabel( "AIR JORDAN Search Count" )
plt.grid(False)

plt.show()
```

```python
import matplotlib.pyplot as plt

# 年ごとのデータと対応するラベル
years = [2020, 2021, 2022, 2023, 2024]
week_data = [
    group_week_2020_AIR_JORDAN["week_number"],
    group_week_2021_AIR_JORDAN["week_number"],
    group_week_2022_AIR_JORDAN["week_number"],
    group_week_2023_AIR_JORDAN["week_number"],
    group_week_2024_AIR_JORDAN["week_number"]
]
count_data = [
    group_week_2020_AIR_JORDAN["AIR JORDAN"],
    group_week_2021_AIR_JORDAN["AIR JORDAN"],
    group_week_2022_AIR_JORDAN["AIR JORDAN"],
    group_week_2023_AIR_JORDAN["AIR JORDAN"],
    group_week_2024_AIR_JORDAN["AIR JORDAN"]
]

# 年ごとにループして棒グラフを作成
for year, week, count in zip(years, week_data, count_data):
    plt.bar(week, count, width=1.0, edgecolor="black")
    plt.xlabel("week number")
    plt.ylabel(f"AIR JORDAN Search count ({year})")
    plt.grid(False)
    plt.title(f"AIR JORDAN Search count - {year}")
    plt.show()
```

```python
import matplotlib.pyplot as plt

# ブランド名のリスト
brands = ["AIR JORDAN", "AIR FORCE", "AIR MAX", "DUNK"]

# 年ごとのデータと対応するラベル
years = [2020, 2021, 2022, 2023, 2024]
week_data = [
    group_week_2020_AIR_JORDAN["week_number"],
    group_week_2021_AIR_JORDAN["week_number"],
    group_week_2022_AIR_JORDAN["week_number"],
    group_week_2023_AIR_JORDAN["week_number"],
    group_week_2024_AIR_JORDAN["week_number"]
]
count_data = [
    group_week_2020_AIR_JORDAN["AIR JORDAN"],
    group_week_2021_AIR_JORDAN["AIR JORDAN"],
    group_week_2022_AIR_JORDAN["AIR JORDAN"],
    group_week_2023_AIR_JORDAN["AIR JORDAN"],
    group_week_2024_AIR_JORDAN["AIR JORDAN"]
]

# 各ブランドに対してループ
for brand in brands:
    # データの取得
    search_2019_brand = Data_2019[["Data", brand, "month", "day", "week_number"]]
    search_2020_brand = Data_2020[["Data", brand, "month", "day", "week_number"]]
    search_2021_brand = Data_2021[["Data", brand, "month", "day", "week_number"]]
    search_2022_brand = Data_2022[["Data", brand, "month", "day", "week_number"]]
    search_2023_brand = Data_2023[["Data", brand, "month", "day", "week_number"]]
    search_2024_brand = Data_2024[["Data", brand, "month", "day", "week_number"]]

    # 年ごとのデータ集計
    group_week_2019_brand = search_2019_brand.groupby("week_number")[brand].sum()
    group_week_2020_brand = search_2020_brand.groupby("week_number")[brand].sum()
    group_week_2021_brand = search_2021_brand.groupby("week_number")[brand].sum()
    group_week_2022_brand = search_2022_brand.groupby("week_number")[brand].sum()
    group_week_2023_brand = search_2023_brand.groupby("week_number")[brand].sum()
    group_week_2024_brand = search_2024_brand.groupby("week_number")[brand].sum()

    # データフレームに変換
    group_week_2020_brand = pd.DataFrame(group_week_2020_brand).reset_index()
    group_week_2021_brand = pd.DataFrame(group_week_2021_brand).reset_index()
    group_week_2022_brand = pd.DataFrame(group_week_2022_brand).reset_index()
    group_week_2023_brand = pd.DataFrame(group_week_2023_brand).reset_index()
    group_week_2024_brand = pd.DataFrame(group_week_2024_brand).reset_index()

    # 週番号と検索数の抽出
    week_data = [
        group_week_2020_brand["week_number"],
        group_week_2021_brand["week_number"],
        group_week_2022_brand["week_number"],
        group_week_2023_brand["week_number"],
        group_week_2024_brand["week_number"]
    ]
    count_data = [
        group_week_2020_brand[brand],
        group_week_2021_brand[brand],
        group_week_2022_brand[brand],
        group_week_2023_brand[brand],
        group_week_2024_brand[brand]
    ]

    # 年ごとにループして棒グラフを作成
    for year, week, count in zip(years, week_data, count_data):
        plt.bar(week, count, width=1.0, edgecolor="black")
        plt.xlabel("week number")
        plt.ylabel(f"{brand} Search count ({year})")
        plt.grid(False)
        plt.title(f"{brand} Search count - {year}")
        plt.show()
```

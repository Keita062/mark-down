```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
california_housing_data = fetch_california_housing()
exp_data = pd.DataFrame(california_housing_data.data, columns=california_housing_data.feature_names)
tar_data = pd.DataFrame(california_housing_data.target, columns=['HousingPrices'])
data = pd.concat([exp_data, tar_data], axis=1)
display(data.head())
```

- Medlnc	世帯所得の中央値
- HouseAge	住宅の築年数
- AveRooms	住宅の部屋数の平均
- AveBedrms	住宅の寝室数の平均
- Poplation	居住人数の合計
- AveOcuup	世帯人数の平均
- Latitude	各地区における代表地区の緯度
- HousePrices	各地区における代表地区の軽度

```python
plt.scatter( data["MedInc"] , data["HousingPrices"] )
plt.xlabel("MedInc")
plt.ylabel("HousingPrices")
plt.show()

np.corrcoef(data["MedInc"], data["HousingPrices"])
```

```python
data[["MedInc","HousingPrices"]].corr()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
X = data[["MedInc"]].values
Y = data["HousingPrices"].values

lr.fit(X, Y)
```

```python
print('coefficient = ', lr.coef_[0]) # 説明変数の係数を出力
print('intercept = ', lr.intercept_) # 切片を出力
```

```python
plt.scatter(X, Y, color = 'blue')         # 説明変数と目的変数のデータ点の散布図をプロット
plt.plot(X, lr.predict(X), color = 'red') # 回帰直線をプロット

plt.xlabel('MedInc')
plt.ylabel('HousingPrices')
plt.xlim(0,18)
plt.ylim(0,6)
plt.grid()

plt.show()                                 # 図の表示
```

```python
lr.score ( X , Y )
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/multiTimeline (6).csv')
```

```python
df.head()
```

```python
# '週'列を日付型に変換
df['week'] = pd.to_datetime(df['Data'])

# 月ごとにグループ化して集計
df['month'] = df['week'].dt.to_period('M')  # 月単位でのグループ化
monthly_data = df.groupby('month')['Searches'].sum()
```

```python
monthly_data
```

```python
# CSVファイルに保存
starbucks_month = monthly_data.to_csv('output.csv')
```

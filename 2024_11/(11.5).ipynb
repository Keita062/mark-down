```python
# 各行で'leftX', 'leftY','rightX', 'rightY'がすべて0であるかを確認
all_zero_mask = (Eye[['leftX', 'leftY','rightX', 'rightY']] == 0).all(axis=1)

group_mask = (all_zero_mask != all_zero_mask.shift()).cumsum()
zero_counts = all_zero_mask.groupby(group_mask).sum()

# 1から5までの連続した0の回数をカウント
results = {}
results[1] = (zero_counts >= 1).sum()
results[2] = (zero_counts >= 2).sum()
results[3] = (zero_counts >= 3).sum()
results[4] = (zero_counts >= 4).sum()
results[5] = (zero_counts >= 5).sum()
results[6] = (zero_counts >= 6).sum()
results[7] = (zero_counts >= 7).sum()
results[8] = (zero_counts >= 8).sum()
results[9] = (zero_counts >= 9).sum()
results[10] = (zero_counts >= 10).sum()

# 結果を表示
print(f'0が1回続いた回数: {results[1]}')
print(f'0が2回続いた回数: {results[2]}')
print(f'0が3回続いた回数: {results[3]}')
print(f'0が4回続いた回数: {results[4]}')
print(f'0が5回続いた回数: {results[5]}')
print(f'0が6回続いた回数: {results[6]}')
print(f'0が7回続いた回数: {results[7]}')
print(f'0が8回続いた回数: {results[8]}')
print(f'0が9回続いた回数: {results[9]}')
print(f'0が10回続いた回数: {results[10]}')
```

```python
# すべての列で0でない行を線形補完
Eye_copy.loc[non_zero_mask] = Eye_copy.loc[non_zero_mask].interpolate()

# 各行でA列、B列、C列、D列がすべて0であるかを確認
all_zero_mask = (Eye_copy[['A', 'B', 'C', 'D']] == 0).all(axis=1)

# すべての列が0である行のインデックスを使って、連続した0をカウントする
group_mask = (all_zero_mask != all_zero_mask.shift()).cumsum()
zero_counts = all_zero_mask.groupby(group_mask).sum()

# 1から5までの連続した0の回数をカウント
results = {}
results[1] = (zero_counts >= 1).sum()
results[2] = (zero_counts >= 2).sum()
results[3] = (zero_counts >= 3).sum()
results[4] = (zero_counts >= 4).sum()
results[5] = (zero_counts >= 5).sum()

# 結果を表示
print(f'0が1回続いた回数: {results[1]}')
print(f'0が2回続いた回数: {results[2]}')
print(f'0が3回続いた回数: {results[3]}')
print(f'0が4回続いた回数: {results[4]}')
print(f'0が5回続いた回数: {results[5]}')
```
